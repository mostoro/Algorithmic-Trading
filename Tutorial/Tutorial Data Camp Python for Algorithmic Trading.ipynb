BUILDING A TRADING STRATEGY WITH PYTHON

A SIMPLE TRADING STRATEGY

```python
short_window = 3
long_window = 10
#Initialize the "signals" dataframe with the "signal" column
signals = pd.DataFrame(index=EURUSD.index)
signals["signal"] = 0.0
#Create short simple moving average over the short window
signals["short_mavg"] = EURUSD["close"].rolling(window
=short_window, min_periods=1, center=False).mean()
#Create long simple moving average over the long window
signals["long_mavg"] = EURUSD["close"].rolling(window
=long_window, min_periods=1, center=False).mean()
#Create signals
signals["signal"][short_window:] = np.where\
(signals["short_mavg"][short_window:]>signals["long_mavg"][short_window:], 1.0, 0.0)
#Generate trading orders
signals["positions"] = signals["signal"].diff()
signals["positions"].fillna(0, inplace=True)
print(signals)
```

PLOT THE SHORT AND LONG MOVING AVERAGES, TOGETHER WITH THE BUY AND SELL SIGNALS

```python
# Initialize the plot figure
fig = plt.figure()
# Add a subplot and label for y-axis
ax1 = fig.add_subplot(111, ylabel="Price in $")
# Plot the closing price
EURUSD["close"].plot(ax=ax1,  color="r", lw=2.)
# Plot the short and long moving averages
signals[["short_mavg", "long_mavg"]].plot(ax=ax1, lw=2.)
# Plot the buy signals
ax1.plot(signals.loc[signals.positions == 1.0].index, signals.short_mavg[signals.positions == 1.0], "^", markersize=10, color="m")
# Plot the sell signals
ax1.plot(signals.loc[signals.positions == -1.0].index, signals.short_mavg[signals.positions == -1.0], "v", markersize=10, color="k")
# Show the plot
plt.show()
```

                          BACKTESTING THE TRADING STRATEGY

```python
signals.index
```

```python
# Set the initial capital
initial_capital = float(100000.0)
# Create a DataFrame `positions`
positions = pd.DataFrame(index=signals.index).fillna(0.0)
# Buy a 1 lot
positions["EURUSD"] = 1*signals["signal"]
# Initialize the portfolio with value owned
portfolio = positions.multiply(EURUSD["close"], axis=0)
# Store the difference in shares owned 
pos_diff = positions.diff()
pos_diff.fillna(0, inplace=True)
# Add `holdings` to portfolio
portfolio["holdings"] = (positions.multiply(EURUSD["close"], axis=0)).sum(axis=1)
# Add `cash` to portfolio
portfolio["cash"] = initial_capital - (pos_diff.multiply(EURUSD["close"], axis=0)).sum(axis=1).cumsum()
# Add `total` to portfolio
portfolio["total"] = portfolio["cash"] + portfolio["holdings"]
# Add `returns` to portfolio
portfolio["returns"] = portfolio["total"].pct_change()
# Print the first lines of `portfolio`
print(portfolio.head())
```

                                      VISUALIZE THE PORTFOLIO

```python
#Create a figure
fig = plt.figure()

ax1 = fig.add_subplot(111, ylabel="Portfolio value in $")
# Plot the equity curve in dollars
portfolio["total"].plot(ax=ax1, lw=2.)

ax1.plot(portfolio.loc[signals.positions == 1.0].index, portfolio.total[signals.positions == 1.0], "^", markersize=10, color="m")
ax1.plot(portfolio.loc[signals.positions == -1.0].index, portfolio.total[signals.positions ==-1.0], "v", markersize=10, color="k")
# Show the plot
plt.show()
```

EVALUATING MOVINGA AVERAGE CROSSOVER STRATEGY

```python
#Isolate the returns of your strategy
returns = portfolio["returns"]
# annualized Sharpe ratio
sharpe_ratio = np.sqrt(252) * (returns.mean() / returns.std())
#Print the sharpe ratio
print(sharpe_ratio)
```

                                MAXIMUM DRAWDOWN

```python
# Define a trailing 252 trading day window
window = 252
# Calculate the max drawdown in the past window days for each day 
rolling_max = EURUSD["close"].rolling(window, min_periods=1).max()
daily_drawdown = EURUSD["close"]/rolling_max - 1.0
# Calculate the minimum (negative) daily drawdown
max_daily_drawdown = daily_drawdown.rolling(window, min_periods=1).min()
# Plot the results
daily_drawdown.plot()
max_daily_drawdown.plot()
#Show the plot
plt.show()
```

                                    COMPOUND ANNUAL GROWTH RATE(CAGR)

```python
# Get the number of days in `EURUSD`
days = (EURUSD.index[-1] - EURUSD.index[0]).days
# Calculate the CAGR 
cagr = ((((EURUSD["close"][-1]) / EURUSD["close"][1])) ** (365.0/days)) - 1
# Print the CAGR
print(cagr)
```
